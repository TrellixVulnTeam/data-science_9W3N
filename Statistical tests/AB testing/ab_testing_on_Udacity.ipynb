<font size="6"><a id='1'>1. Introduction</a></font> 

I will be conducting A/B test on the data collected from Udacity website to make recommendation. 

An experiment was conducted in which potential Udacity students were diverted by cookie into two groups, experiment and control. The experiment group was asked to input the amount of time they are willing to devote to study, after clicking a "start free trial button", whereas the control group was not. Three invariant metrics (Number of Cookies, Number of clicks on "start free trial", and Click-Through-Probability) were chosen for purposes of validation and sanity checking while Gross Conversion (enrollment/cookie) and Net Conversion (payments/cookie) served as evaluation metrics. The null hypothesis is that there is no difference in the evaluation metrics between the groups, futhermore, a practical signifcance threshold was set for each metric.


```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

control = pd.read_csv(filepath_or_buffer = "/home/micah/Desktop/data/udacity_ab_data/control_data.csv")
experiment = pd.read_csv(filepath_or_buffer = "/home/micah/Desktop/data/udacity_ab_data/experiment_data.csv")
```

<font size="6"><a id='2'>2. A/B testing</a></font> 


```python
control = control.dropna()
experiment = experiment.dropna()
#Pageviews  Clicks  Enrollments  Payments

control['CP_C'] = control['Clicks']/control['Pageviews']
control['EP_C'] = control['Enrollments']/control['Pageviews']
control['PP_C'] = control['Payments']/control['Pageviews']
control['EC_C'] = control['Enrollments']/control['Clicks']
control['PC_C'] = control['Payments']/control['Clicks']



experiment['CP_E'] = experiment['Clicks']/experiment['Pageviews']
experiment['EP_E'] = experiment['Enrollments']/experiment['Pageviews']
experiment['PP_E'] = experiment['Payments']/experiment['Pageviews']
experiment['EC_E'] = experiment['Enrollments']/experiment['Clicks']
experiment['PC_E'] = experiment['Payments']/experiment['Clicks']


#print control.head()
#print experiment.head()
Pageviews_c =  control['Pageviews'].sum()
Clicks_c = control['Clicks'].sum()
Enrollments_c = control['Enrollments'].sum()
Payments_c = control['Payments'].sum() 

Pageviews_e = experiment['Pageviews'].sum()
Clicks_e = experiment['Clicks'].sum()
Enrollments_e = experiment['Enrollments'].sum()
Payments_e = experiment['Payments'].sum()
```

```python
from statsmodels.stats.proportion import proportions_ztest
count = np.array([Clicks_c, Clicks_e])
count2 = np.array([Enrollments_c, Enrollments_e])
count3 = np.array([Payments_c, Payments_e])

# default option is  alternative='two-sided'
nobs = np.array([Pageviews_c, Pageviews_e])
print proportions_ztest(count, nobs)
print proportions_ztest(count2, nobs)
print proportions_ztest(count3, nobs)
print proportions_ztest(count2, count)
```

```python
fig = plt.figure(figsize=(20,8))
#control = control.set_index('Date')
#print control#['CP_C']
CP = pd.concat([control['CP_C'], experiment['CP_E'] ], axis=1)
#print CP
#df1 = pd.concat([control.set_index('Date'),experiment.set_index('Date')], axis=1, join='inner').reset_index()
CP.columns = ['Control', 'Experiment']
ax = CP.plot(lw=2, colormap='jet', marker='.', markersize=10, title='Clicks over pageview ratio')

ax.set_xlabel("days since the experiment started")
ax.set_ylabel("ratio")

plt.show()
print 'Z, P values: ',proportions_ztest(count, nobs)

```

```python
fig = plt.figure(figsize=(20,8))

EP = pd.concat([control['EP_C'], experiment['EP_E'] ], axis=1)
#print CP
#df1 = pd.concat([control.set_index('Date'),experiment.set_index('Date')], axis=1, join='inner').reset_index()
EP.columns = ['Control', 'Experiment']
ax = EP.plot(lw=2, colormap='jet', marker='.', markersize=10, title='Enrollment over pageview ratio')

ax.set_xlabel("days since the experiment started")
ax.set_ylabel("ratio")

plt.show()
print 'Z, P values: ', proportions_ztest(count2, nobs)

```

```python
fig = plt.figure(figsize=(20,8))

PP = pd.concat([control['PP_C'], experiment['PP_E'] ], axis=1)
#print CP
#df1 = pd.concat([control.set_index('Date'),experiment.set_index('Date')], axis=1, join='inner').reset_index()
PP.columns = ['Control', 'Experiment']
ax = PP.plot(lw=2, colormap='jet', marker='.', markersize=10, title='Payment over pageview ratio')

ax.set_xlabel("days since the experiment started")
ax.set_ylabel("ratio")

plt.show()

print 'Z, P values: ', proportions_ztest(count3, nobs)
```

```python
# EC_C
fig = plt.figure(figsize=(20,8))

EC = pd.concat([control['EC_C'], experiment['EC_E'] ], axis=1)
#print CP
#df1 = pd.concat([control.set_index('Date'),experiment.set_index('Date')], axis=1, join='inner').reset_index()
EC.columns = ['Control', 'Experiment']
ax = EC.plot(lw=2, colormap='jet', marker='.', markersize=10, title='Gross conversion: Enrollment over click ratio')

ax.set_xlabel("days since the experiment started")
ax.set_ylabel("ratio")

plt.show()
print 'Gross conversion: Z, P values: ', proportions_ztest(count2, count)
```

```python
# EC_C
fig = plt.figure(figsize=(20,8))

PC = pd.concat([control['PC_C'], experiment['PC_E'] ], axis=1)
#print CP
#df1 = pd.concat([control.set_index('Date'),experiment.set_index('Date')], axis=1, join='inner').reset_index()
PC.columns = ['Control', 'Experiment']
ax = PC.plot(lw=2, colormap='jet', marker='.', markersize=10, title='Net conversion: Payment over click ratio')

ax.set_xlabel("days since the experiment started")
ax.set_ylabel("ratio")

plt.show()
print 'Net conversion: Z, P values: ', proportions_ztest(count3, count)
```
