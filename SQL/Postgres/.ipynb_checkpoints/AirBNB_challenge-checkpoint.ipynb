Questions: 
    1. What's the most expensive listing? What else can you tell me about the listing?
    2. What neighborhoods seem to be the most popular?
    3. What time of year is the cheapest time to go to San Francisco? What about the busiest?

Answers:
    1. SELECT t1.calender_date, t1.available, t1.price, t2.* FROM (select listing_id, calender_date, available, price FROM sfo_calendar WHERE price!='' ORDER BY price DESC limit 1) t1 JOIN sfo_listings t2 ON t1.listing_id = t2.id;
    
    -- We had to integrate the following queries: 
    select listing_id, calender_date, available, price from sfo_calendar where price!='' order by price DESC limit 1; 
    select * from sfo_listings order by price DESC limit 1;
   
   
    2. SELECT avg(reviews_per_month), neighbourhood FROM sfo_listings WHERE reviews_per_month>0  GROUP BY neighbourhood ORDER BY avg(reviews_per_month) DESC;
