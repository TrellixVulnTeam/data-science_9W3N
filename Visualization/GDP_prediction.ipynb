<font size="6"><a id='4'>4. Data Imputation</a></font> 


For simplicity, We would just fill the missing data using the median of the region that a country belongs, as countries that are close geologically are often similar in many ways. However, the mean and the median GDP per capita for the some same region differs greatly. For this reason, we will not only use region similarity to fill in the missing data but also use economic similarity(GDP per capita).
To find out the economic similarity, we will be creating bin for all the countires based on the GDP per capita

```python
df.groupby('Region')[['GDP ($ per capita)','Literacy (%)','Agriculture']].median()
```

```python
df.groupby('Region')[['GDP ($ per capita)','Literacy (%)','Agriculture']].mean()
```

```python
# GDP per capita for Western Sahara is missing. So this is my guestimate based on the regional average. 
#df.loc[df.Country=='Western Sahara', 'GDP ($ per capita)'] = 1000
df['Region'] = df['Region'].map(lambda x: x.strip())
df['Country'] = df['Country'].map(lambda x: x.strip())

df['GDP_bucket'] = pd.cut(df['GDP ($ per capita)'], [0,1005,3955,12235, 60000], labels=["low", "middle", "upper", "high"])

#df = df.drop([223])
#print df.loc[df.Region=='NORTHERN AFRICA']


#print df.head(n=6)

for col in df.columns.values:
    if df[col].isnull().sum() == 0:
        continue
    if col == 'Climate':
        guess_values = df.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
    else:
        #guess_values = df.groupby('Region')[col].median()
        try:
            guess_values = df.groupby(['Region','GDP_bucket'])[col].median()
                        #guess_values = df.groupby('Region')[col].median()

        except:
            pass
            #guess_values1 = df.groupby('Region')[col].median()
        #guess_values = data.groupby('Region')[col].median()
        guess_values1 = df.groupby('Region')[col].median()



    # try to assign according to its region and gdp bucket, otherwise just use region     
    for region in df['Region'].unique():    
        for bucket in df['GDP_bucket'].unique(): 
            #if df[col].loc[df[col].isnull()]:   
            try:
                df[col].loc[(df[col].isnull())&(df['Region']==region) &(df['GDP_bucket']==bucket)] = guess_values[region][bucket] 
            except:
                pass
            df[col].loc[(df[col].isnull())&(df['Region']==region)] = guess_values1[region]

         
        
print df.isna().sum()


```

<font size="6"><a id='6'>6. Feature Importance</a></font> 


Let find out the features that are most predictive powers to a target variable GDP per capita

```python
from xgboost import XGBRegressor
from xgboost import plot_importance
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X = df[['Population', 'Area (sq. mi.)',
       'Pop. Density (per sq. mi.)', 'Coastline (coast/area ratio)',
       'Net migration', 'Infant mortality (per 1000 births)', 'Literacy (%)', 'Phones (per 1000)', 'Arable (%)',
       'Crops (%)', 'Other (%)', 'Climate', 'Birthrate', 'Deathrate',
       'Agriculture', 'Industry', 'Service']]

y = df['GDP ($ per capita)']
y = y.convert_objects(convert_numeric=True)
print (y.dtypes)
print (X[['Pop. Density (per sq. mi.)','Coastline (coast/area ratio)','Net migration']].head())
dims = (15, 10)
fig, ax = plt.subplots(figsize=dims)


# fit model no training data
model = XGBRegressor()
model.fit(X, y)
plot_importance(model, ax = ax)
plt.show()
# feature importance
print(model.feature_importances_)
```

<font size="6"><a id='7'>7. GDP Prediction</a></font> 


```python
# Prediction
model1 = XGBRegressor()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)

model1.fit(X_train, y_train)
# make predictions for test data and evaluate
train_pred_Y = model.predict(X_train)
y_pred = model1.predict(X_test)

predictions = [round(value) for value in y_pred]

from sklearn.metrics import r2_score
print ('r square: ',r2_score(y_test, predictions) )

from sklearn.feature_selection import SelectFromModel


thresholds = [0.06, 0.07]
for thresh in thresholds:
    selection = SelectFromModel(model1, threshold=thresh, prefit=True)
    select_X_train = selection.transform(X_train)
    selection_model = XGBRegressor()
    selection_model.fit(select_X_train, y_train)
    select_X_test = selection.transform(X_test)
    y_pred = selection_model.predict(select_X_test)
    predictions = [round(value) for value in y_pred]
    print ('r square: ',r2_score(y_test, predictions) )
```
