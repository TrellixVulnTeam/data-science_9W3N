```python
fig, axes = plt.subplots(nrows=10, ncols=3, figsize=(20,70))
plt.subplots_adjust(hspace=0.4)


# Lets find the correlation between the target variable and the numeric variables

corr_to_price = pd.Series()
for col in num_feats:
    corr_to_price[col] = train['SalePrice'].corr(train[col])   

corr_to_price = corr_to_price.sort_values(ascending=False)
corr_to_price = corr_to_price.round(decimals=2)


for i in range(10):
    for j in range(3):

        try: 
           #print (corr_to_price.index.values[i*3+j])
           sns.regplot(x=corr_to_price.index.values[i*3+j], y='SalePrice', data=train,ax=axes[i,j], fit_reg=False, marker='.')
           title = 'correlation='+str(corr_to_price[i*3+j])
           axes[i,j].set_title(title)
        except:
           pass

#axes[1,2].set_xlim(0,102)
plt.show()


def outliers(x, y): 
   # Set figure width to 12 and height to 9
   fig_size = [4, 4]
   plt.rcParams["figure.figsize"] = fig_size

   fig, ax = plt.subplots()
   #fig, ax = plt.subplot(, 3, 1)

   ax.scatter(x = train[x], y = train[y])
   plt.ylabel(y, fontsize=10)
   plt.xlabel(x, fontsize=10)
   plt.show()

   return ''
```


<font size="6"><a> Skewness</a></font> 

Here we can take 2 different approaches. First we take skew() built in function from dataframe, then we can get skew() function from Scipy. It is interesting to see that Scipy actually misses some of the variables due to the way it handles missing values differently.


```python
# 1st approach
skews= data[num_feats].skew().sort_values()
skews = skews[skews> 1]
skews = skews.sort_values(ascending=False)
print (skews)

# 2nd approach
from scipy import stats

skews1 = data[num_feats].apply(lambda x: stats.skew(x))
skews1 = skews1[skews1> 1]
skews1 = skews1.sort_values(ascending=False)
print (skews1)
```
